```python
import requests
import json
from pprint import pprint
import pandas as pd
import numpy as np
import os
import csv
import matplotlib.pyplot as plt

```

```python
query_url = 'https://api.ed.gov/data/mbk-labor-force-participation?'
api_key = 'api_key=bVIf4sU0IzjYfs473ZQEyzOjmZ5hkWeG6vdk5vPc'
per_page = 'per_page=' + '1258'
full_url = query_url + api_key + "&" + per_page

labour_data = requests.get(full_url).json()
pprint(labour_data)
```

```python
#count of data
labour_len = len(labour_data['resources'])
labour_len
```

```python
#column lists
labour_percentage = []
labour_year = []
labour_race = []
labour_sex = []

#for loop to create columns
for a in range(labour_len):
    labour_percentage.append(labour_data['resources'][a]['Percentage'])
    labour_year.append(int(labour_data['resources'][a]['Year']))
    labour_race.append(labour_data['resources'][a]['Race/ethnicity'])
    labour_sex.append(labour_data['resources'][a]['Sex'])
    
#dataframe

labour_race_df = pd.DataFrame({
    "Year" : labour_year,
    "Race" : labour_race,
    "Percentage" : labour_percentage
})

labour_sex_df = pd.DataFrame({
    "Year" : labour_year,
    "Sex" : labour_sex,
    "Percentage" : labour_percentage
})
```

```python
#dropna for both dataframes

labour_race_df = labour_race_df.dropna(how='any')
labour_sex_df = labour_sex_df.dropna(how='any')

labour_race_df['Percentage'] = labour_race_df['Percentage'].astype(float)
labour_sex_df['Percentage'] = labour_sex_df['Percentage'].astype(float)

#labour_race_df
#labour_sex_df
```

```python
#drop years 1980-1999

labour_race_df_2 = labour_race_df[labour_race_df['Year']>=2000]
#labour_race_df_2
labour_sex_df_2 = labour_sex_df[labour_sex_df['Year']>=2000]
#labour_sex_df_2
```

```python
#replace Asian race to other
labour_race_df_2 = labour_race_df_2.replace(to_replace='Asian', value='Other')
labour_race_df_2 = labour_race_df_2.replace(to_replace=['White', 'Black'], value=['Caucasian', 'African-American'])
#labour_race_df_2
```
