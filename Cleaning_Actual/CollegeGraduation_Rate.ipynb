```python
#import Dependencies

import requests
import json
from pprint import pprint
import pandas as pd
import numpy as np
import os
import csv
import matplotlib.pyplot as plt
```

```python
#Get Requests

query_url = 'https://api.ed.gov/data/mbk-bachelors-or-higher?'
api_key = 'api_key=bVIf4sU0IzjYfs473ZQEyzOjmZ5hkWeG6vdk5vPc'
per_page = 'per_page=' + '360'
full_url = query_url + api_key + "&" + per_page

CG_data = requests.get(full_url).json()

pprint(CG_data)
```

```python
#length
length = len(CG_data['resources'])
```

```python
#collect data and loop through list

percentage=[]
year=[]
sex=[]
race=[]

for a in range(length):
    year.append(int(CG_data['resources'][a]['Year']))
    sex.append(CG_data['resources'][a]['Sex'])
    race.append(CG_data['resources'][a]['Race/ethnicity'])
    percentage.append(CG_data['resources'][a]['Percentage'])
```

```python
#create dataframes for race and gender

CG_race = pd.DataFrame({'Year': year, 'Race': race, 'Percentage': percentage})
CG_sex = pd.DataFrame({'Year': year, 'Sex': sex, 'Percentage': percentage})
#CG_race['Race'].unique()

CG_sex.head()
```

```python
#replace race values to other
CG_race2 = CG_race.replace(to_replace=['Asian, non-Hispanic', 'Pacific Islander, non-Hispanic','American Indian/Alaska Native, non-Hispanic', 'Two or more races, non-Hispanic'], value = 'Other')
CG_race3 = CG_race2.replace(to_replace=['White, non-Hispanic', 'Black, non-Hispanic'], value=['Caucasian', 'African-American'])
#CG_race3['Race'].unique()       

#replace plural sexes to aggregate counts
CG_sex = CG_sex.replace(to_replace=['Females', 'Males'], value=['Female', 'Male'])
#CG_sex['Sex'].unique()
```

```python
#drop nones, convert percentage to float

CG_race3 = CG_race3.dropna(how='any')
CG_sex = CG_sex.dropna(how='any')

CG_race3['Percentage'] = CG_race3['Percentage'].astype(float)
CG_sex['Percentage'] = CG_sex['Percentage'].astype(float)
```

```python
#Groupby per race and Year, then unstack to create Race columns and get percentages by year.
CG_race4 = CG_race3.groupby(['Year', 'Race'])
CG_race5 = CG_race4.mean().unstack(level=-1).stack(level=0)
CG_race5.reset_index(level=0, inplace=True)
CG_race5 = CG_race5.set_index('Year')
CG_race5.plot(kind='line')
```

```python
#Groupby per race and Year, then unstack to create Race columns and get percentages by year.
CG_sex1 = CG_sex.groupby(['Year', 'Sex'])
CG_sex2 = CG_sex1.mean().unstack(level=-1).stack(level=0)
CG_sex2.reset_index(level=0, inplace=True)
CG_sex2 = CG_sex2.set_index('Year')
CG_sex2.plot(kind='line')
```
